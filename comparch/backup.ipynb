## Αριθμητική λογική μονάδα (ALU)

```scala
class Alu(word_w: Int) extends Module {
  val io = IO(new Bundle {
    val a = Input(UInt(word_w.W))
    val b = Input(UInt(word_w.W))
    val sel = Input(UInt(2.W))
    val sub = Input(UInt(1.W))
    val out = Output(UInt(word_w.W))
    val c = Output(UInt(1.W))
  })
  
  io.c := 0.U
  io.out := 0.U
  when (io.sel==="b00".U) {
      io.out := io.a & io.b
  }.elsewhen(io.sel==="b01".U) {
      io.out := io.a | io.b
  }.elsewhen(io.sel==="b10".U) {
      io.out := io.a ^ io.b
  }.otherwise {  
      val sum = Wire(UInt((word_w+1).W))
      when (io.sub===0.U) {
          sum := io.a +& io.b
      }.otherwise {
          sum := io.a -& io.b
      }
      io.out := sum(word_w-1,0)
      io.c := sum(word_w)
  } 
}
```

## Σύνολο καταχωρητών (register file)

```scala
class RegisterFile(word_w: Int, sel_w: Int) extends Module {
  
  val io = IO(new Bundle {
    val in = Input(UInt(word_w.W))
    val read_sel_a = Input(UInt(sel_w.W))
    val read_sel_b = Input(UInt(sel_w.W))
    val write_sel = Input(UInt(sel_w.W))      
    val out_a = Output(UInt(word_w.W))
    val out_b = Output(UInt(word_w.W))    
  })
    
  val reg_num = scala.math.pow(2,sel_w).toInt

    
  val regFile = RegInit(VecInit(Seq.fill(reg_num)(0.U(word_w.W))))
  
  when(io.write_sel=/=0.U) {
    regFile(io.write_sel) := io.in
  }
  io.out_a := regFile(io.read_sel_a)
  io.out_b := regFile(io.read_sel_b)
    
}
```
