```python
from simpleast import ASTNode
```

```python
a = ASTNode(attributes={'type':'NUMBER','value':3.14})
b = ASTNode(attributes={'type':'NUMBER','value':5.0})
c = ASTNode(subnodes=[a,b],attributes={'type':'OP','func':'+'})
d = ASTNode(attributes={'type':'NUMBER','value':11.23})
e = ASTNode(subnodes=[c,d],attributes={'type':'OP','func':'*'})
```

```python
print(e)
```

```python
print(e.attribute('type'))
```

```python
print(e.subnode(0))
```

```python
def walk(node):
    if node.attribute('type')=='NUMBER':
        print("I'm a number with value {}".format(node.attribute('value')))
    
    else:
        print("I'm an operator with function {}".format(node.attribute('func')))
        # walk subnodes
        walk(node.subnode(0))
        walk(node.subnode(1))
```

```python
walk(e)
```
