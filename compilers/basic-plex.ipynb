# Βασική χρήση του module plex
Ξεκινήστε κάνοντας import το module plex:

```python
import plex
```

Στη συνέχεια προσδιορίστε τα patterns που θα αναγνωρίζονται (εδώ, τα strings `abc` και `123`, καθώς και διάφοροι χαρακτήρες whitespace):

```python
text_abc = plex.Str('abc')
text_123 = plex.Str('123')
spaces = plex.Rep1(plex.Any(' \t\n'))
```

Κατασκευάστε το `lexicon` που περιέχει τα tokens που θα επιστραφούν όταν αναγνωρίζεται κάθε pattern (ή άλλες ενέργειες):

```python
lexicon = plex.Lexicon([
    (text_abc,'ABC_TOKEN'),
    (text_123,'123_TOKEN'),
    (spaces,plex.IGNORE)
    ])
```

Στο σημείο αυτό πρέπει να κάνετε upload ένα δοκιμαστικό αρχείο κειμένου από το site του μαθήματος (`plex-123-abc.txt`).

Αμέσως μετά εκτελέστε το επόμενο για να κάνετε λεκτική ανάλυση του κειμένου:

```python
with open("plex-123-abc.txt","r") as fp:

    scanner = plex.Scanner(lexicon,fp)

    while True:
        token,lexeme = scanner.read()

        if not token: break    # reached end-of-text (EOT)
    
        print(token,lexeme)
```

Δοκιμάστε να εισάγετε χαρακτήρες που δεν αναγνωρίζονται από το `lexicon`. Εκτελέστε το προηγούμενο κελί και παρατηρήστε το σφάλμα που παράγεται.

Μπορούμε να τροποποιήσουμε τον κώδικα ώστε να "πιάνουμε" το σφάλμα και να εκτυπώνουμε τη θέση που συνέβει:

```python
with open("plex-123-abc.txt","r") as fp:

    scanner = plex.Scanner(lexicon,fp)

    while True:
        
        try:
            token,lexeme = scanner.read()
        except plex.errors.PlexError:
            _,lineno,charno = scanner.position()  # lineno is 1-based, charno is 0-based
            print("Scanner Error at line {} char {}".format(lineno,charno+1))
            break    # lexical analysis ends after error
        if not token: break    # reached end-of-text (EOT)
    
        print(token,lexeme)
```
