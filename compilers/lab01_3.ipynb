# Python και Κανονικές Εκφράσεις
Βρείτε παρακάτω τα βήματα για τη χρήση των κανονικών εκφράσεων στην Python.

## 1. Χρήση της βιβλιοθήκης re
Η βιβλιοθήκη των κανονικών εκφράσεων `re` ανήκει στις standard βιλβιοθήκες της Python. Έτσι δεν χρειάζεται η εγκατάσταση third-party libraries. Απλά θα πρέπει να γίνει το import της βιβλιοθήκης **μια φορά μόνο, στην αρχή του προγράματος**.

Εκτελέστε το ακόλουθο πριν συνεχίσετε:

```python
import re
```

## 2. Δημιουργία αντικειμένου κανονικής έκφρασης
Στο επόμενο βήμα δημιουργούμε ένα αντικείμενο κανονικής έκφρασης. Το αντικείμενο αυτό μας παρέχει τις μεθόδους αναζήτησης. Η δημιουργία γίνεται **μία φορά**. Στη συνέχεια μπορεί το αντικείμενο να χρησιμοποιηθεί **πολλές φορές** κατά τη διάρκεια εκτέλεσης του προγράμματος.

Εκτελέστε το επόμενο για να δημιουργήσετε ένα αντικείμενο κανονικής έκφρασης με όνομα μεταβλητής `rexp`:

```python
rexp = re.compile(r'^cat')
```

*Σημ.: η σύνταξη `r'...'` εισάγει ένα raw string, όπου οι ακολουθίες με escape όπως το `\n` δεν μετατρέπονται αλλά παραμένουν ξεχωριστοί χαρακτήρες. Αυτό δεν είναι αναγκαίο στο συγκεκριμένο παράδειγμα αλλά θα χρειαστεί πολλές φορές όταν η ίδια η κανονική έκφραση χρησιμοποιεί escape sequences.*

## 3. Αναζήτηση μέσω κανονικής έκφρασης
Εδώ χρησιμοποιούμε τη μέθοδο `search()` που βρίσκει το πρώτο ταίριασμα μέσα στο κείμενο (αν υπάρχει): 

```python
m = rexp.search('category')
```

Αν υπάρχει ταίριασμα, επιστρέφεται ένα match object με την πληροφορία ταιριάσματος.

Δοκιμάστε να τυπώσετε το `m`:

```python
m
```

Αν δεν υπάρχει ταίριασμα το `m` είναι `None`:

```python
m = rexp.search('black cat')
m
```

## 4. Χρήση στη λογική του προγράμματός μας
Παράδειγμα:

```python
m = rexp.search('category')
if m:
    print(m.group(0))
else:
    print('Not found')
```

To `m.group(0)` μας παρέχει το κείμενο που ταίριαξε η κανονική έκφραση.
